```python
import json



def load_file(file_path):

    """

    :param file_path: path to the json file

    :return: an array in which each entry is tuple [review, classification label]

    """

    with open(file_path) as json_file:

        raw_data = json.load(json_file)

        return convert_data(raw_data)
        # return raw_data



def convert_data(raw_data):

    data = []

    for elem in raw_data:

        data.append([elem["comment"], elem["label"]])

    return data
```

```python
def clean_text(df):
  df['text'] = df['text'].str.replace("\r", " ")
  df['text'] = df['text'].str.replace("\n", " ")
  df['text'] = df['text'].str.replace('"', '')
  df['text'] = df['text'].str.lower()
  return df

df_bug = clean_text(df_bug)
df_rating = clean_text(df_rating)
df_feature = clean_text(df_feature)
df_experience = clean_text(df_experience)
```

```python
# remove punctuations

def remove_puncts(df):
  punctuation_signs = list("?:!.,;")

  for punct_sign in punctuation_signs:
      df['text'] = df['text'].str.replace(punct_sign, ' ', regex=False)

  df['text'] = df['text'].str.replace("'s", "", regex=True)
  return df

df_bug = remove_puncts(df_bug)
df_rating = remove_puncts(df_rating)
df_feature = remove_puncts(df_feature)
df_experience = remove_puncts(df_experience)
```

```python
# lemmatization

from nltk.stem import WordNetLemmatizer

def lemmatize(df):
  wordnet_lemmatizer = WordNetLemmatizer()
  nrows = len(df)
  lemmatized_text_list = []

  for index, row in df.iterrows():  # Use iterrows() to correctly iterate over DataFrame rows
      # Create an empty list which will contain the lemmatized words
      lemmatized_list = []

      # Save the text and its words into an object
      text = row['text']
      text_words = text.split(" ")

      # Iterate through every word to lemmatize
      for word in text_words:
          lemmatized_list.append(wordnet_lemmatizer.lemmatize(word, pos="v"))

      # Join the list
      lemmatized_text = " ".join(lemmatized_list)

      # Append to the list containing the texts
      lemmatized_text_list.append(lemmatized_text)

  # Set the 'text' column in df to the list of lemmatized texts AFTER the loop
  df.loc[:,'text'] = lemmatized_text_list
  return df
```

```python
df_bug = lemmatize(df_bug)
df_rating = lemmatize(df_rating)
df_feature = lemmatize(df_feature)
df_experience = lemmatize(df_experience)
```

```python
# remove stop words function

def remove_stop_words(df):
  for stop_word in stop_words:
    regex_stopword = r"\b" + stop_word + r"\b"
    df.loc[:, 'text'] = df['text'].str.replace(regex_stopword, '', regex=True)
  return df
```

```python
# tf-idf

from sklearn.feature_extraction.text import TfidfVectorizer

def fit_tfidf(X_train):
  ngram_range = (1,2)
  min_df = 0.01 # ignore terms that appear in less than 1% of the documents
  max_df = 0.7 # ignore terms that appear in more than 80% of the documents
  max_features = 300

  tfidf = TfidfVectorizer(encoding='utf-8',
                          ngram_range=ngram_range,
                          stop_words=None,
                          lowercase=False,
                          max_df=max_df,
                          min_df=min_df,
                          max_features=max_features,
                          norm='l2',
                          sublinear_tf=True)

  # Fit tf-idf vectorizer
  tfidf_vectorizer = tfidf.fit(X_train)

  return tfidf_vectorizer
```

Train SVM model with Random Search hyperparameter tuning

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
import numpy as np

def train_svm_model(features, labels, label_column='label'):
    le = LabelEncoder()
    encoded_labels = le.fit_transform(labels)
    svm = SVC()

    # hyperparameter options
    C = [.0001, .001,  .1, 1]
    gamma = [.0001, .001, .01, .1, 1, 10, 100]
    degree = [1, 2, 3, 4, 5]
    kernel = ['linear', 'rbf', 'poly']
    probability = [True]

    # Create the param grid
    param_grid = {'C': C,
                  'kernel': kernel,
                  'gamma': gamma,
                  'degree': degree,
                  'probability': probability
                }

    # Setup the random search with cross-validation
    random_search = RandomizedSearchCV(svm,
                                       param_distributions=param_grid,
                                       n_iter=50,
                                       scoring='f1',
                                       cv=5,
                                       verbose=1,
                                       random_state=42,
                                       )

    # Fit the random search model
    random_search.fit(features, encoded_labels)
    

    print("Best parameters found: ", random_search.best_params_)
    print("Best cross-validated score achieved: ", random_search.best_score_)

    best_svc = random_search.best_estimator_
    return best_svc.fit(features, encoded_labels)
    
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, accuracy_score, recall_score, precision_score
from scipy.stats import mode

def aggregate_predictions(classifier_mappings, predictions):
    label_predictions = [[] for _ in range(len(predictions['bug_rating']))]

    # Aggregate predictions for each classifier
    for key, pred in predictions.items():
        class1, class2 = classifier_mappings[key]
        for i, p in enumerate(pred):
            # Map binary prediction back to original labels
            actual_label = class1 if p == 0 else class2
            label_predictions[i].append(actual_label)

    # Determine the most common label for each instance
    final_predictions = [mode(preds)[0] for preds in label_predictions]
    return final_predictions

classifier_mappings = {
    'bug_rating': (1, 2),
    'bug_feature': (1, 3),
    'bug_experience': (1, 4),
    'rating_feature': (2, 3),
    'rating_experience': (2, 4),
    'experience_feature': (3, 4)
}
```

```python
final_predictions = aggregate_predictions(classifier_mappings, predictions)
```
